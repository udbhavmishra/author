```python
import pytesseract
from PIL import Image
import pdf2image
import numpy as np

# Function to extract text from PDF using OCR
def extract_text_from_pdf(pdf_path):
    # Convert PDF to list of PIL images
    pages = pdf2image.convert_from_path(pdf_path)

    all_text = []

    for page in pages:
        # Convert page to grayscale
        img_gray = page.convert('L')

        # Use Tesseract OCR to extract text
        text = pytesseract.image_to_string(img_gray)

        all_text.append(text)

    return all_text

# Function to detect columns in extracted text
def detect_columns(text):
    # Split text by lines
    lines = text.splitlines()

    # Determine column boundaries (example: using space count)
    columns = []
    current_column = []
    for line in lines:
        if line.strip():  # Non-empty line
            current_column.append(line.strip())
        else:  # Empty line, indicates column break
            if current_column:
                columns.append(current_column)
                current_column = []

    # Add last column if any lines left
    if current_column:
        columns.append(current_column)

    return columns

# Example usage
if __name__ == "__main__":
    # Path to your PDF file
    pdf_path = "/content/AUT1-20240530-124_eml_merged.pdf"

    # Extract text from PDF
    extracted_text = extract_text_from_pdf(pdf_path)

    # Combine all text extracted from pages
    combined_text = " ".join(extracted_text)

    # Detect columns
    columns = detect_columns(combined_text)

    # Print columns (you may want to further process or save them)
    for i, column in enumerate(columns):
        print(f"Column {i+1}:")
        for line in column:
            print(line)
        print("\n")
```
