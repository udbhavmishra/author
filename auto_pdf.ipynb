```python
import fitz  # PyMuPDF
import pytesseract
import cv2
import pandas as pd
import re
from PIL import Image
import numpy as np
import io

# Specify the path to tesseract executable if it's not in your PATH
pytesseract.pytesseract.tesseract_cmd = r'/usr/local/bin/tesseract'  # Update with your Tesseract path

# Function to convert PDF to images
def pdf_to_images(pdf_path):
    doc = fitz.open(pdf_path)
    images = []
    for page_num in range(len(doc)):
        page = doc.load_page(page_num)
        pix = page.get_pixmap()
        img = Image.open(io.BytesIO(pix.tobytes()))
        images.append(img)
    return images

# Function to preprocess image for better OCR
def preprocess_image(image):
    gray = cv2.cvtColor(np.array(image), cv2.COLOR_BGR2GRAY)
    _, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)
    return thresh

# Function to extract text using Tesseract OCR
def extract_text(image):
    text = pytesseract.image_to_string(image, config='--psm 6')
    return text

# Function to extract vehicle data from text
def extract_vehicle_data(text):
    pattern = re.compile(r'(\d{4})\s+([\w\s]+)\s+\$\s+-\s+(\w+)\s+([\w\d]+)\s+([\w\s]+)\s+\d+')
    matches = pattern.findall(text)
    data = []
    for match in matches:
        year, make_model, lpn, serial_number, territory = match
        data.append({
            "Year": year,
            "Make & Model": make_model,
            "Serial Number": serial_number,
            "Territory": territory
        })
    return data

# Main function to process the PDF and extract data
def process_pdf(pdf_path):
    images = pdf_to_images(pdf_path)
    all_data = []
    for img in images:
        preprocessed_img = preprocess_image(img)
        text = extract_text(preprocessed_img)
        data = extract_vehicle_data(text)
        all_data.extend(data)
    return all_data

# Path to the PDF file
pdf_path = '"C:\Users\UdbhavMishra\Desktop\AUT1-20240530-122_eml_merged.pdf"'

# Process the PDF and extract data
vehicle_data = process_pdf(pdf_path)

# Convert the extracted data to a DataFrame and display
df = pd.DataFrame(vehicle_data)
print(df)

# Optionally, save the data to a CSV file
df.to_csv('vehicle_data.csv', index=False)
```

```python
import fitz  # PyMuPDF
import pytesseract
import cv2
import pandas as pd
import re
from PIL import Image
import numpy as np
import io

# Specify the path to tesseract executable if it's not in your PATH
pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'  # Update with your Tesseract path

# Function to convert PDF to images
def pdf_to_images(pdf_path):
    doc = fitz.open(pdf_path)
    images = []
    for page_num in range(len(doc)):
        page = doc.load_page(page_num)
        pix = page.get_pixmap()
        img = Image.open(io.BytesIO(pix.tobytes()))
        images.append(img)
    return images

# Function to preprocess image for better OCR
def preprocess_image(image):
    gray = cv2.cvtColor(np.array(image), cv2.COLOR_BGR2GRAY)
    _, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)
    return thresh

# Function to extract text using Tesseract OCR
def extract_text(image):
    text = pytesseract.image_to_string(image, config='--psm 6')
    return text

# Function to extract vehicle data from text
def extract_vehicle_data(text):
    pattern = re.compile(r'(\d{4})\s+([\w\s&]+)\s+\$\s+-\s+([\w\d]+)\s+([\w\d]+)\s+([\w\s]+)\s+\d+')
    matches = pattern.findall(text)
    data = []
    for match in matches:
        year, make_model, lpn, serial_number, territory = match
        data.append({
            "Year": year,
            "Make & Model": make_model,
            "Serial Number": serial_number,
            "Territory": territory
        })
    return data



# Main function to process the PDF and extract data
def process_pdf(pdf_path):
    try:
        images = pdf_to_images(pdf_path)
    except Exception as e:
        print(f"Error converting PDF to images: {e}")
        return []

    all_data = []
    for img in images:
        try:
            preprocessed_img = preprocess_image(img)
            text = extract_text(preprocessed_img)
            print("Extracted Text: ", text)  # Print the extracted text for verification
            data = extract_vehicle_data(text)
            all_data.extend(data)
        except Exception as e:
            print(f"Error processing image: {e}")

    return all_data

# Path to the PDF file
pdf_path = r"/content/AUT1-20240530-122_eml_merged.pdf"

# Process the PDF and extract data
vehicle_data = process_pdf(pdf_path)

# Convert the extracted data to a DataFrame and display
df = pd.DataFrame(vehicle_data)
print(df)
  # Remove blank columns
df = df.dropna(axis=1, how='all')
df.head()


# # Optionally, save the data to a CSV file
# df.to_csv('vehicle_data.csv', index=False)
```

```python
import PyPDF2
import pandas as pd

# Read the PDF
pdf_path = '/content/AUT1-20240530-122_eml_merged.pdf'
reader = PyPDF2.PdfFileReader(pdf_path)
text = ''
for page_num in range(reader.numPages):
    page = reader.getPage(page_num)
    text += page.extract_text()

# Manually parsing the text to extract the table content
# Assuming the table starts with "NO. UNIT YEAR MAKE & MODEL LPN SERIAL NUMBER TERRITORY CLASS/USE"
import re

pattern = re.compile(r"(\d+ - +\d{4} +\w+ +\w+ +\w+ +\S+ +\S+ +Alberta +\d+)")
matches = pattern.findall(text)

# Extracted data into a list
data = []
for match in matches:
    # Split the match string by spaces, handling multiple spaces and cleaning
    parts = re.split(r'\s{2,}', match)
    data.append(parts)

# Convert to DataFrame
columns = ["NO.", "UNIT", "YEAR", "MAKE & MODEL", "LPN", "SERIAL NUMBER", "TERRITORY", "CLASS/USE"]
df = pd.DataFrame(data, columns=columns)

# Remove blank columns
df = df.dropna(axis=1, how='all')
df.head()

# # Save to CSV
# csv_path = '/content/AUT1-20240530-122_eml_merged.pdf'
# df.to_csv(csv_path, index=False)
# csv_path
```
