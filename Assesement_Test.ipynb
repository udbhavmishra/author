##Load the CSV data into a Pandas DataFrame:


```python
import pandas as pd

# Load the data from the CSV file
csv_file_path = '/content/Sample_Sales_Data.csv'  # Update the path if needed
sales_data = pd.read_csv(csv_file_path)

# Display the first few rows of the DataFrame
print(sales_data.head())
```

##Calculate the total sales for each product category:

```python
# Calculate the total sales for each product category
total_sales_by_category = sales_data.groupby('ProductCategory')['TotalSales'].sum().reset_index()

# Display the result
print("Total Sales by Product Category:")
print(total_sales_by_category)
```

##Find the top 5 products with the highest sales:

```python
top_5_products_sort = sales_data.sort_values(by='TotalSales', ascending=False).head()
print("\nTop 5 Products with the Highest Sales (sort_values):")
print(top_5_products_sort)
```

5.	Numpy Array Operations

–	Create a 3x3 NumPy array with random integers. Perform the following operations:
•	Compute the sum of all elements in the array.
•	Find the mean and standard deviation of the elements.
•	Normalize the array (subtract the mean and divide by the standard deviation).


```python
# Step 1: Create a 3x3 NumPy array with random integers
import numpy as np
array = np.random.randint(1, 100, size=(3, 3))
print("Original Array:")
print(array)
```

```python
# Step 2: Compute the sum of all elements in the array
array_sum = np.sum(array)
print("\nSum of all elements:", array_sum)
```

```python
# Step 3: Find the mean and standard deviation of the elements
array_mean = np.mean(array)
array_std = np.std(array)
print("Mean of elements:", array_mean)
print("Standard Deviation of elements:", array_std)
```

```python
# Step 4: Normalize the array (subtract the mean and divide by the standard deviation)
normalized_array = (array - array_mean) / array_std
print("\nNormalized Array:")
print(normalized_array)
```
